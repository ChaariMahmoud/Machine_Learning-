```python
films_series_tunisie = data[data['country'].str.contains('Tunisia', na=False)]
```

```python
print(len(films_series_tunisie))
```

```python
films_series_france = data[data['country'].str.contains('France', na=False)]
```

```python
print("Nombre de films/séries produits par la France :", len(films_series_france))
```

```python
contenu_plus_vieux = data[data['release_year'] == data['release_year'].min()]
print("Contenu le plus vieux disponible sur Netflix :\n", contenu_plus_vieux)
```

```python
contenus_plus_anciens = data.sort_values(by='release_year').head(5)
print("Les 5 contenus les plus anciens :\n", contenus_plus_anciens)
```

```python
directeurs_plus_productifs = data['director'].value_counts().head(5)
print("Directeurs ayant produit le plus de films/séries disponibles sur Netflix :\n", directeurs_plus_productifs)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
data['type'].value_counts().plot(kind='bar')
plt.title('Nombre de films/séries')
plt.xlabel('Type')
plt.ylabel('Nombre')
plt.show()
```

```python
top_pays_producteurs = data['country'].value_counts().head(10)
print("La France figure-t-elle parmi les 10 premiers pays producteurs de films/séries ?")
print("France" in top_pays_producteurs)
```

```python
plt.figure(figsize=(10, 6))
data[data['type'] == 'TV Show']['duration'].str.extract('(\d+)', expand=False).astype(float).value_counts().sort_index().plot(kind='bar')
plt.title('Nombre de séries par nombre de saisons')
plt.xlabel('Nombre de saisons')
plt.ylabel('Nombre de séries')
plt.show()
```
